# Normalisation de la Matrice X(features)

```python
 
def normalisation(x):
    x_norme=x.copy()
    moy=np.zeros(x.shape[1])
    sig=np.zeros(x.shape[1])
    n=x_norme.shape[1]
    sig=[np.std(x[:,0]),np.std(x[:,1]),np.std(x[:,2]),np.std(x[:,3])]
    moy=[np.mean(x[:,0]),np.mean(x[:,1]),np.mean(x[:,2]),np.mean(x[:,3])]
    for i in range(n) :
            x_norme[:,i]=(x_norme[:,i] - moy[i])/sig[i]
        
    return  x_norme,moy,sig
    
```

# Teste et recuperation des valeurs de la fonction de normalisation

```python
x_norme,moy,sig= normalisation(x)
x_norme
x=x_norme.copy()
```

 # Definition du MODELE

# Parametre F= X.O

```python
# x est ' une matrice' que l'on doit concatener avec une colone rempli de 1
X=np.hstack((x,np.ones((x.shape[0],1))))
X
```

```python
def model(X,O):
    return X.dot(O)
```

```python

X.shape
```

```python
# calcul des theta
O=np.random.randn(5,1)
O
```

```python
# teste du modele
model(X,O)
```

# Fonction COUT :j(O)=1/2msomme(X.O-Y)^2

```python
def fonction_cout(X,O,y):
    m=len(y)
    return 1/2*m*np.sum((model(X,O)-y)**2)
```

```python
# teste de la fonction
fonction_cout(X,O,y)# retourne toujours une valeur
```

# GRADIENT dj(O)/d(O)=1/m*XT.(X.O-y)

```python
def grad(X,O,y):
    m=len(y)
    return 1/m*X.T.dot(model(X,O)-y)
```

```python
# teste de la fonction
grad(X,O,y)
```

```python
# observation du dataset
plt.scatter(x[:,1],y)
```

# GRADIENT DESCENTE

```python
def gradient_descente(X,O,y,learning_rate,n_iterations):
            cost_history=np.zeros(n_iterations)
            for i in range(0,n_iterations):
                           O=O-learning_rate*grad(X,O,y)
                           cost_history[i]= fonction_cout(X,O,y)
            return O, cost_history
```

# Machine Learning

```python
theta_final,cost_history=gradient_descente(X,O,y,learning_rate=0.01,n_iterations=1000)
print(theta_final,cost_history)
```

```python
predictions=model(X,theta_final)
```

#  courbe du nombre d'iterations

```python
# courbe du nombre d'iterations
plt.plot(range(1000), cost_history,c='r',label='cost_history')
```
